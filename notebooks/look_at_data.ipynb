```python
import sys

sys.path.insert(0, "../src")

import matplotlib.pyplot as plt  # noqa: E402

from mlops_project.data import MyDataset  # noqa: E402
```

```python
# Load dataset (limit to 100 samples for quick exploration)
dataset = MyDataset(data_path="../data", limit=100)

print(f"Dataset length: {len(dataset)}")
```

```python
# Look at annotations
print("Annotations shape:", dataset.annotations.shape)
print("\nColumns:", dataset.annotations.columns.tolist())
print("\nFirst 5 rows:")
dataset.annotations.head()
```

```python
# Label distribution
print("Label distribution:")
print(dataset.annotations["label"].value_counts())
```

```python
# Load and display a sample image
image, label = dataset[0]

print(f"Image size: {image.size}")
print(f"Image mode: {image.mode}")
print(f"Label: {label}")

plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.title(f"Label: {label}")
plt.axis("off")
plt.show()
```

```python
# Display grid of sample images
fig, axes = plt.subplots(2, 4, figsize=(12, 6))

for i, ax in enumerate(axes.flat):
    image, label = dataset[i]
    ax.imshow(image)
    ax.set_title(f"Label: {label}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```
